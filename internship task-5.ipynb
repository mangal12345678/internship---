```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
target = pd.Series(iris.target, name='species')

df = pd.concat([data, target], axis=1)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data)
df_pca = pd.DataFrame(data=pca_result, columns=['principal component 1', 'principal component 2'])
df_pca['species'] = target
df_pca
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_pca['principal component 1'], kde=True)
plt.title('Histogram of Principal Component 1')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df_pca['principal component 2'],kde=True)
plt.title('histogram of principal component 2')
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='principal component 1',y='principal component 2',hue='species',data=df_pca,palette='viridis',s=80)
plt.title("scatter plot of principal components")
plt.show
```
